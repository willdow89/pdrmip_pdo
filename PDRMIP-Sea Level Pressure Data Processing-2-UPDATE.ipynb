Regridding models so they have same lat/long divisions as hadgem2 (highest latitude resolution) 

```python
canesm2_b_psl_rg = canesm2_b_psl.regrids(hadgem2_b_psl, method = 'bilinear')
mpiesm_b_psl_rg = mpiesm_b_psl.regrids(hadgem2_b_psl, method = 'bilinear')
noresm1_b_psl_rg = noresm1_b_psl.regrids(hadgem2_b_psl, method = 'bilinear')
cesm1cam4_b_psl_rg = cesm1cam4_b_psl.regrids(hadgem2_b_psl, method = 'bilinear')
cesm1cam5_b_psl_rg =  cesm1cam5_b_psl.regrids(hadgem2_b_psl, method = 'bilinear')
sprintars_b_psl_rg = sprintars_b_psl.regrids(hadgem2_b_psl, method = 'bilinear')
hadgem2_b_psl_rg = hadgem2_b_psl
#hadgem3_b_psl_rg = hadgem3_b_psl.regrids(hadgem2_b_psl, method = 'bilinear')
giss_b_psl_rg = giss_b_psl.regrids(hadgem2_b_psl, method = 'bilinear')
ipsl_b_psl_rg = ipsl_b_psl.regrids(hadgem2_b_psl, method = 'nearest_stod')
echam_b_psl_rg = echam_b_psl.regrids(hadgem2_b_psl, method = 'bilinear')
```

Transposing so all fields have same shape

```python
canesm2_b_psl_rg_tp = canesm2_b_psl_rg.transpose(['T','Y','X'])
mpiesm_b_psl_rg_tp = mpiesm_b_psl_rg.transpose(['T','Y','X'])
noresm1_b_psl_rg_tp = noresm1_b_psl_rg.transpose(['T','Y','X']) 
cesm1cam4_b_psl_rg_tp = cesm1cam4_b_psl_rg.transpose(['T','Y','X']) 
cesm1cam5_b_psl_rg_tp = cesm1cam5_b_psl_rg.transpose(['T','Y','X'])
sprintars_b_psl_rg_tp = sprintars_b_psl_rg.transpose(['T','Y','X']) 
hadgem2_b_psl_rg_tp = hadgem2_b_psl_rg.transpose(['T','Y','X'])
#hadgem3_b_psl_rg_tp = hadgem3_b_psl_rg.transpose(['T','Y','X'])
giss_b_psl_rg_tp = giss_b_psl_rg.transpose(['T','Y','X']) 
ipsl_b_psl_rg_tp = ipsl_b_psl_rg.transpose(['T','Y','X']) 
echam_b_psl_rg_tp = echam_b_psl_rg.transpose(['T','Y','X']) 
```
